```python
import pandas as pd
```

```python
df = pd.read_csv('canada-data.csv')
df.rename(columns={'Postcode': 'PostalCode'}, inplace=True)
df
```

```python
ndf = df[df['Borough'] != 'Not assigned']
ind = range(0, 210)
ndf.index = ind
ndf
```

```python
pdf = ndf.astype(str).groupby('PostalCode', sort=False).agg(lambda x: ','.join(x.unique()))
pdf.insert(loc=0,column='PostalCode',value=pdf.index)
pdf.index = range(0, pdf.shape[0])
pdf
```

```python
pdf.loc[pdf['Neighbourhood'] == 'Not assigned', 'Neighbourhood'] = pdf.loc[pdf['Neighbourhood'] == 'Not assigned', 'Borough'].iloc[0]
pdf
```
